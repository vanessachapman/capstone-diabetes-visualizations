# Data Visualizations for Diabetes Prediction Application

```python
# Import all the tools we need

# Regular exploratory data analysis and plotting libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## Load Data

```python
df = pd.read_csv("diabetes.csv")
df.rename(columns = {'Diabetes_binary' : 'target'}, inplace=True)
df.head()
```

## Diabetes Frequency According to Sex

```python
# Create a plot of crosstab
pd.crosstab(df.target, df.Sex).plot(kind="bar",
                                   figsize=(10, 6),
                                   color=["salmon", "lightblue"])
plt.title("Diabetes Frequency According to Sex")
plt.xlabel("0 = No Diabetes, 1 = Diabetes/Pre-Diabetes")
plt.ylabel("Amount")
plt.legend(["Female", "Male"])
plt.xticks(rotation=0);
```

## Age vs BMI for Diabetes

```python
# Create another figure
plt.figure(figsize=(10, 6))

# Scatter with positive examples
plt.scatter(df.Age[df.target==1],
           df.BMI[df.target==1],
           c="salmon")

# Scatter with negative examples
plt.scatter(df.Age[df.target==0],
           df.BMI[df.target==0],
           c="lightblue")

# Add labels and helpful information
plt.title("Diabetes in Function of Age and BMI")
plt.xlabel("Age")
plt.ylabel("BMI")
plt.legend(["Diabetes/Pre-Diabetes", "No Diabetes"]);
```

## Diabetes Frequency per General Health

```python
pd.crosstab(df.GenHlth, df.target).plot(kind="bar",
                                       figsize=(10,6),
                                       color=["salmon", "lightblue"]);

# Add labels and helpful information
plt.title("Diabetes Frequency per General Health")
plt.xlabel("General Health Score")
plt.ylabel("Amount")
plt.legend(["No Diabetes", "Diabetes/Pre-Diabetes"])
plt.xticks(rotation=0);
```

## Diabetes Frequency per High Blood Pressure

```python
pd.crosstab(df.HighBP, df.target).plot(kind="bar",
                                      figsize=(10, 6),
                                      color=["salmon", "lightblue"])

# Add labels
plt.title("Diabetes Frequency for High Blood Pressure Status")
plt.xlabel("High Blood Pressure Status (0 = No High BP, 1 = High BP)")
plt.ylabel("Amount")
plt.legend(["No Diabetes", "Diabetes/Pre-Diabetes"])
plt.xticks(rotation=0);
```

## Correlation Matrix

```python
corr_matrix = df.corr()
fig, ax = plt.subplots(figsize=(20, 15))
ax = sns.heatmap(corr_matrix,
                annot=True,
                linewidth=0.5,
                fmt=".2f",
                cmap="YlGnBu");
```
